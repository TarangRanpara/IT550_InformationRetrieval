```python
import os
os.chdir('drive/MyDrive/IRLAB/A1/data')
data_files = os.listdir()
```



**1.   Read and combine the text between the <TEXT> tags by using Beautiful Soup from all the documents.**


```python
from bs4 import BeautifulSoup
data = []
for file in data_files:
  #reading each file using bs4
  with open(file) as fp:
    soup = BeautifulSoup(fp, 'html.parser')
    data.append(soup.text)

#joining data
data = "".join(data)
data
```


**2.   Remove punctuation and numerical values from your text.**





```python
import re

#remove numbers
data = re.sub(r'\d', '', data)

#remove punctuation
data = re.sub(r'[^\w\s\d]', '', data)

data
```



**3. Perform tokenization on the text. There are two ways to perform tokenization here. One way is to split
feature of string and another is using NLTK library. Display the difference between both approaches.**

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# using nltk word tokenizer
nltk_tokens = word_tokenize(data)
```

```python
# using conventional string split for generating tokens
split_tokens = data.split()
```

```python
# comparing both results
print('NLTK_Tokens'.ljust(20), 'Split_Tokens'.rjust(28))
print(''.ljust(50, '-'))
for n, s in zip(nltk_tokens, split_tokens):
  print(n.ljust(20), '<----->', s.rjust(20))
```

**4. Convert the text to lowercase**

```python
data = data.lower()
data
```

**5. Remove stop-words from the text**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# set of stop words
stop_words = set(stopwords.words('english')) 
tokens = word_tokenize(data)

# removing stop words
filtered_tokens = [token for token in tokens if not token in stop_words]

filtered_tokens[:30]
```

**6. Perform Stemming (use Porter Stemmer) and Lemmatization (separately) and display the differences that
you observe while using the two approaches. (For this you have to use NLTK library)**

```python
from nltk.stem import PorterStemmer 

ps = PorterStemmer()
stemmed_words = list(map(ps.stem, filtered_tokens))
```

```python
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
  
lemmatizer = WordNetLemmatizer() 

lemmatized_words = list(map(lemmatizer.lemmatize, filtered_tokens))
```

```python
# comparing stemming and lemmztizing
print('Stemming'.ljust(20), 'Lemmatizing'.rjust(28))
print('-'*50)
for s, l in zip(stemmed_words, lemmatized_words):
  print(s.ljust(20), '<----->', l.rjust(20))
```
